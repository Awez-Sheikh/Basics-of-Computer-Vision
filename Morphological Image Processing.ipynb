```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
img=cv2.imread("/content/tumer.jpg")
print(img.shape)
print(img)
```

```python
cv2_imshow(img)
```

```python
#apply binary thresholding
ret,binary_img=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
#define kernal
kernal=np.ones((5,5),np.uint8)
#apply dilation
dilation=cv2.dilate(binary_img,kernal,iterations=1)
#eroded image
eroded_image=cv2.erode(dilation,kernal,iterations=1)
```

```python
openned_image=cv2.morphologyEx(eroded_image,cv2.MORPH_OPEN,kernal)
closed_image=cv2.morphologyEx(openned_image,cv2.MORPH_CLOSE,kernal)
gradient_image=cv2.morphologyEx(closed_image,cv2.MORPH_GRADIENT,kernal)

```

```python
cv2_imshow(gradient_image)
```

```python
tophat_image=cv2.morphologyEx(gradient_image,cv2.MORPH_TOPHAT,kernal)
blackhat_image=cv2.morphologyEx(tophat_image,cv2.MORPH_BLACKHAT,kernal)
```

```python
titles=['Threshold Image','Dialated Image','Eroded Image','Opened Image','Closed Image','Gradient Image','Tophat Image','Blackhat Image']
```

```python
images = [binary_img, dilation,eroded_image, openned_image, closed_image, gradient_image]

for i in range(6):
    plt.figure(figsize=(5,5))
    plt.imshow(images[i], cmap='grey')
    plt.axis('off')
    plt.title(titles[i])
    plt.show()
```
